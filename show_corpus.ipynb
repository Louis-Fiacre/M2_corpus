# Scrapper Gallica

### imports, fonctions et constantes

```python
import os
import sys
import re
import urllib.request, urllib.error, urllib.parse
from bs4 import BeautifulSoup
```

```python
def mkdir(chemin_dossier):
        """ créer un dossier dans un chemin s'il n'existe pas """
        if not os.path.exists(chemin_dossier):
            os.mkdir(chemin_dossier)
        return(chemin_dossier)
```

```python
def import_meta(dossier, ark, name='metadonnees.xml'):
        """télécharge depuis gallica le xml dublin core des metadonnees d'un id ark"""
        
        metadonnee_path = os.path.join(dossier, name)

        if not os.path.isfile(metadonnee_path):
            url = 'https://gallica.bnf.fr/services/OAIRecord?ark='+ark[7:]
            try:
                urllib.request.urlretrieve(url, metadonnee_path)
            except urllib.error.URLError as error:
                print(error)

        return metadonnee_path
```

```python
def extract_meta(path_xml_file):
    """ cherche dans balise des métadonnées d'un ouvrage gallica 
    si une version océrisée de l'ouvrage existe et nombre de vue """
    
    if path_xml_file.endswith(".xml"):

        xml_file = open(path_xml_file, 'r')
        doc = xml_file.read()
        soup = BeautifulSoup(doc, 'lxml')

        ocr = soup.find('nqamoyen') # balise responsable de l'ocr
        ocr = ocr.contents
        ocr = float(ocr[0])
        
        dcformat = soup.find_all('dc:format') # balise responsable du nombre de vue
        for balise in dcformat:

            if balise.contents[0][0:24] == 'Nombre total de vues :  ':
                vues = balise.contents[0].replace('Nombre total de vues :  ','')

        if ocr >= 50:

            return (ocr,vues)

        return (False,vues)
```

### Nombre de vues et taux ocr de gallica

```python
df_gallica = df_num.dropna(axis=0, subset=['Gallica'])
del df_gallica['Genre']
del df_gallica['Wikisource']
del df_gallica['Autres_sources']
del df_gallica['Numérisé']
del df_gallica['Date_redaction']

liste_ocr = []
liste_nb_vues = []

for i in df_gallica.index:
    namedossier = df_gallica['Auteurs'][i]+'_'+df_gallica['Titres'][i]
    dossier = os.path.join(CWD,namedossier)
    mkdir(dossier)
    #print(namedossier)
    
    # import métadonnées en xml
    import_meta(dossier,df_gallica['Gallica'][i])
    
    # extrait taux erreurs ocr et nombre de vues
    ocr, nb_vues = extract_meta(os.path.join(dossier,'metadonnees.xml'))
    #print('  OCR_Gallica:',ocr, '\n  Vues:', nb_vues)
    
    liste_ocr.append(ocr)
    liste_nb_vues.append(nb_vues)

df_gallica['OCR_Gallica']=liste_ocr
df_gallica['Nb_vues_Gallica']=liste_nb_vues
df_gallica
    
```

### Scrappe ocr si minimum à 85% ou images si ocr est inférieur ou inexistant

```python
def scrapper(ark, nb_vues, dossier, ext):

    """ 
    récupère les images ou l'xml d'un ouvrage depuis gallica 
    et les dépose dans le répertoire correspondant
    """
    nb_vues = int(nb_vues)
    listpage = range(0, nb_vues+1)
    len_last = len(str(nb_vues))

    for page in listpage:

        size = len_last - len(str(page))
        zero = "0" * size

        file_ = zero + str(page) + ext 

        path_file = os.path.join(dossier, file_)
        error = ""
        if not os.path.isfile(path_file):

            if ext == '.jpg':
                # requête pour les jpg
                url = 'http://gallica.bnf.fr/iiif/ark:' + ark + '/f' + str(page) + '/full/5000/0/native.jpg'
            
            if ext == '.xml':
            # requête pour les xml
                url = 'https://gallica.bnf.fr/RequestDigitalElement?O='+ark[7:]+'&E=ALTO&Deb='+str(page)
            
            try:

                urllib.request.urlretrieve(url, path_file)
                

            except urllib.error.HTTPError as err:

                error = ('Error -> ({})'.format(err))
                print(error,'\n',url)
                
                
        # les lignes suivantes gèrent juste un affichage pour suivre l'evolution des requêtes
        print_path = dossier.replace('/home/lf/Documents/Master/Memoire/corpus/','')
        try:
            percent = round((page / nb_vues)*100)
        except:
            print(print_path, page, nb_vues)
            
        sys.stdout.write("\r"+ print_path + " : " + str(percent) + "% " + zero+str(page)+"/"+str(nb_vues)+ext)
        sys.stdout.flush()
    sys.stdout.write("\r\n")
```

```python
for i in df_gallica.index:
    namedossier = df_gallica['Auteurs'][i]+'_'+df_gallica['Titres'][i]
    dossier = os.path.join(CWD,namedossier)
    
    
    if str(df_gallica['Gallica'][i])!='/12148/bpt6k2056392':

        if df_gallica['OCR_Gallica'][i] >= 85:
            xmldossier = mkdir(os.path.join(dossier,'gallica_ocr_xml'))
            scrapper(df_gallica['Gallica'][i],df_gallica['Nb_vues_Gallica'][i], xmldossier, '.xml')
        else:
            jpgdossier = mkdir(os.path.join(dossier,'gallica_img'))
            scrapper(df_gallica['Gallica'][i],df_gallica['Nb_vues_Gallica'][i], jpgdossier, '.jpg')
```

```python
def xml_alto_to_txt(xml_path, txt_path):

    """ 
    Transforme les xml téléchargés depuis gallica en plein texte txt
    
    f : nom du fichier xml
    xml_dir : dossier du fichier xml
    txt_dir : dossier du fichier txt créé"""

    files = sorted(os.listdir(xml_path))

    firstpage = int(files[0][0:-4])
    lastpage = int(files[-1][0:-4])

    for f in files:

        if f.endswith(".xml"):

            xml_file = os.path.join(xml_path,f)
            ftxt = f.replace(".xml",".txt")
            txt_file = os.path.join(txt_path,ftxt)

            if not os.path.isfile(txt_file):


                fd = open(xml_file, 'r')
                doc = fd.read()
                fd.close()

                soup = BeautifulSoup(doc, 'lxml')
                strin = soup.find_all('string')
                content = ""

                for data in strin:
                
                    content+=data['content']+" "

                with open(txt_file,'w') as txtfile:
                    txtfile.write(content)

        print_path = txt_path.replace('/home/lf/Documents/Master/Memoire/Corpus/','')
        percent = round((int(f[0:-4]) / lastpage)*100)
        sys.stdout.write("\r"+ print_path + " : " + str(percent) + "% " + str(f[0:-4])+"/"+str(files[-1].replace(".xml",'.txt')))
        sys.stdout.flush()
    sys.stdout.write("\r\n\n")
```
